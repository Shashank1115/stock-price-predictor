```python
import pandas as pd

df = pd.read_csv("Cleaned_Dataset1.csv")  # Use correct file name here
df['Date'] = pd.to_datetime(df['Date'])

df.head()
```

```python
def create_sequences(data, seq_length=60):
    X = []
    y = []
    for i in range(seq_length, len(data)):
        X.append(data[i-seq_length:i])
        y.append(data[i])
    return np.array(X), np.array(y)
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tqdm.notebook import tqdm
import os


# Create a directory to save models
output_dir = "stock_models"
os.makedirs(output_dir, exist_ok=True)

symbols = df['Series'].unique()
model_summaries = []

for symbol in tqdm(symbols, desc="Training models"):
    stock_df = df[df['Series'] == symbol].sort_values('Date')

    if len(stock_df) < 100:
        continue  # Skip short datasets

    # Get only Close price
    close_prices = stock_df[['Close']].values
    scaler = MinMaxScaler()
    scaled_prices = scaler.fit_transform(close_prices)

    # Build sequences
    X, y = create_sequences(scaled_prices)
    X = X.reshape((X.shape[0], X.shape[1], 1))

    # Train-test split
    split = int(0.8 * len(X))
    X_train, y_train = X[:split], y[:split]
    X_test, y_test = X[split:], y[split:]

    # Build LSTM model
    model = Sequential([
        LSTM(50, return_sequences=True, input_shape=(X.shape[1], 1)),
        LSTM(50),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mean_squared_error')
    model.fit(X_train, y_train, epochs=5, batch_size=32, verbose=0)

    # Save model
    model_file = os.path.join(output_dir, f"{symbol}_model.h5")
    model.save(model_file)

    model_summaries.append({
        "symbol": symbol,
        "model_path": model_file,
        "samples": len(stock_df)
    })

print(" Models saved successfully!")
```

```python
def predict_stock(stock_name, model_path, csv_path="Cleaned_Dataset1.csv", days_to_predict=126):
    import numpy as np
    import pandas as pd
    from keras.models import load_model
    from sklearn.preprocessing import MinMaxScaler
    import matplotlib.pyplot as plt

    # Load model
    model = load_model(model_path)

    # Load and filter data
    df = pd.read_csv(csv_path)
    df = df[df["Series"] == stock_name]
    close_prices = df["Close"].values.reshape(-1, 1)

    # Normalize
    scaler = MinMaxScaler()
    scaled_close = scaler.fit_transform(close_prices)

    # Prepare last sequence
    sequence_length = 60
    last_sequence = scaled_close[-sequence_length:]
    predicted_scaled = []
    current_input = last_sequence.copy()

    for _ in range(days_to_predict):
        prediction = model.predict(current_input.reshape(1, sequence_length, 1), verbose=0)
        predicted_scaled.append(prediction[0][0])
        current_input = np.append(current_input[1:], [[prediction[0][0]]], axis=0)

    # Inverse transform
    predicted_prices = scaler.inverse_transform(np.array(predicted_scaled).reshape(-1, 1))

    # Plot
    plt.figure(figsize=(12, 6))
    plt.plot(range(len(close_prices)), close_prices, label='Historical')
    plt.plot(range(len(close_prices), len(close_prices) + days_to_predict), predicted_prices, label='Predicted', color='orange')
    plt.title(f"{stock_name} Closing Price Prediction (Next {days_to_predict} Days)")
    plt.xlabel("Days")
    plt.ylabel("Closing Price (INR)")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    return predicted_prices
```

```python
def plot_next_60_days(stock_name, model_path):
    from keras.models import load_model
    from sklearn.preprocessing import MinMaxScaler
    import pandas as pd
    import numpy as np
    import matplotlib.pyplot as plt

    model = load_model(model_path)

    df = pd.read_csv("Cleaned_Dataset1.csv")
    df = df[df["Series"] == stock_name]

    close_prices = df["Close"].values.reshape(-1, 1)

    scaler = MinMaxScaler()
    scaled_prices = scaler.fit_transform(close_prices)

    sequence_length = 60
    last_sequence = scaled_prices[-sequence_length:]

    predicted_scaled = []
    current_input = last_sequence.copy()

    for _ in range(60):  # Next 60 trading days
        prediction = model.predict(current_input.reshape(1, sequence_length, 1), verbose=0)
        predicted_scaled.append(prediction[0][0])
        current_input = np.append(current_input[1:], [[prediction[0][0]]], axis=0)

    predicted_prices = scaler.inverse_transform(np.array(predicted_scaled).reshape(-1, 1))

    plt.figure(figsize=(12, 6))
    plt.plot(range(len(close_prices)), close_prices, label='Historical')
    plt.plot(range(len(close_prices), len(close_prices) + 60), predicted_prices, label='Predicted (Next 60 Days)', color='orange')
    plt.title(f"{stock_name} Closing Price Prediction (Next 60 Days)")
    plt.xlabel("Days")
    plt.ylabel("Price (INR)")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
def predict_next_60_days(stock_name, model_path):
    from keras.models import load_model
    from sklearn.preprocessing import MinMaxScaler
    import pandas as pd
    import numpy as np

    model = load_model(model_path)

    df = pd.read_csv("Cleaned_Dataset1.csv")
    df = df[df["Series"] == stock_name]

    close_prices = df["Close"].values.reshape(-1, 1)

    scaler = MinMaxScaler()
    scaled_prices = scaler.fit_transform(close_prices)

    sequence_length = 60
    last_sequence = scaled_prices[-sequence_length:]

    predicted_scaled = []
    current_input = last_sequence.copy()

    for _ in range(60):
        prediction = model.predict(current_input.reshape(1, sequence_length, 1), verbose=0)
        predicted_scaled.append(prediction[0][0])
        current_input = np.append(current_input[1:], [[prediction[0][0]]], axis=0)

    predicted_prices = scaler.inverse_transform(np.array(predicted_scaled).reshape(-1, 1))

    # Print values
    print(f"Predicted closing prices for {stock_name} (Next 60 days):")
    for i, price in enumerate(predicted_prices):
        print(f"Day {i+1}: {price[0]:.2f}")

    # Save to CSV
    future_dates = pd.date_range(start=pd.to_datetime(df["Date"].iloc[-1]) + pd.Timedelta(days=1), periods=60, freq='B')
    prediction_df = pd.DataFrame({"Date": future_dates, "Predicted Price": predicted_prices.flatten()})
    prediction_df.to_csv(f"{stock_name}_60_days_prediction.csv", index=False)
    print(f"Predicted prices saved as {stock_name}_60_days_prediction.csv")
```
